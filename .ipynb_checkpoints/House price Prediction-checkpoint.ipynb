```python
def missing_counts(datasets):
    total=dataset1.isna().sum().sort_values(ascending=False)
    percent=(dataset1.isnull().sum()/len(dataset1)).sort_values(ascending=False)
    missing=pd.concat((total,percent),axis=1, keys=['Total','percent'])
    return missing[missing['Total']>0]
```

```python
train=dataset1.iloc[:1460]
test=dataset1.iloc[1460:2919]
```

```python
train_x=train.drop('SalePrice',axis=1)
train_y=train['SalePrice']
test=test.drop('SalePrice',axis=1)
```

```python
x_train=train_x.copy()
x_test=test.copy()
y_train=train_y.copy()
```

```python
print(x_train.shape)
print(x_test.shape)
print(y_train.shape)
```

```python
from sklearn.preprocessing import StandardScaler
sc=StandardScaler()
x_train=sc.fit_transform(x_train)
x_test=sc.fit_transform(x_test)
```

```python
from sklearn.decomposition import PCA
pca=PCA(n_components=0.99)
x_train=pca.fit_transform(x_train)
x_test=pca.transform(x_test)
variance=pca.explained_variance_ratio_
```

```python
from xgboost import XGBRegressor
regressor=XGBRegressor(base_score=0.25, booster='gbtree', colsample_bylevel=1,
       colsample_bytree=1, gamma=0, learning_rate=0.1, max_delta_step=0,
       max_depth=2, min_child_weight=1, missing=None, n_estimators=900,
       n_jobs=1, nthread=None, objective='reg:linear', random_state=0,
       reg_alpha=0, reg_lambda=1, scale_pos_weight=1, seed=None,
       silent=True, subsample=1)
```

```python
regressor.fit(x_train,y_train)
```

```python
y_pred=regressor.predict(x_test)
pd.DataFrame({'Id':list(range(1461,2920)),'SalePrice':y_pred}).to_csv('submission_upd.csv',index=False)
```
